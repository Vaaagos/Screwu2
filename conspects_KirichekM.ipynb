В Python можно выполнять все основные арифметические операции, такие как сложение, вычитание, умножение и деление. Кроме того, Python предлагает ряд математических функций, таких как возведение в степень, извлечение корня, округление числа, вычисление модуля и другие. Вот несколько примеров:

- Сложение:

  ```
  a = 5
  b = 3
  c = a + b
  print(c)
  ```
  Вывод: 8

- Вычитание:

  ```
  a = 5
  b = 3
  c = a - b
  print(c)
  ```
  Вывод: 2

- Умножение:

  ```
  a = 5
  b = 3
  c = a * b
  print(c)
  ```
  Вывод: 15

- Деление:

  ```
  a = 5
  b = 3
  c = a / b
  print(c)
  ```
  Вывод: 1.6666666666666667

- Возведение в степень:

  ```
  a = 5
  b = 3
  c = a ** b
  print(c)
  ```
  Вывод: 125

- Извлечение корня:

  ```
  import math
  
  a = 25
  b = math.sqrt(a)
  print(b)
  ```
  Вывод: 5.0

- Округление числа:

  ```
  a = 1.234567
  b = round(a, 3)
  print(b)
  ```
  Вывод: 1.235

- Вычисление модуля:

  ```
  a = -5
  b = abs(a)
  print(b)
  ```
  Вывод: 5



# Теперь ответим на вторую часть вопроса и укажем три отличия от классической математики:

1. В Python нет неопределенности. Например, результат деления на ноль будет ошибкой (ZeroDivisionError), а не бесконечностью или неопределенностью.

2. В Python можно работать с комплексными числами, а классическая математика не всегда учитывает комплексную составляющую.

3. При работе с большими числами в Python не возникает проблем с переполнением, так как тип данных можно выбрать в зависимости от нужд (например, использовать long для больших целых чисел). В классической математике часто используются формулы и методы, которые могут приводить к переполнению и другим проблемам с точностью.

````python
def bubble_sort(lst):
    n = len(lst)
    for i in range(n-1):
        for j in range(n-i-1):
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]

lst = [3, 14, 1, 7, 9, 8, 11, 6, 4, 2]
bubble_sort(lst)
print(lst)

````
